```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plot
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Train.csv')
data.describe()
```

```python
data.head()
```

```python
data.plot()
```

```python
data.iloc()
```

```python
data['decade'] = 10 * (data['YearBuilt'] // 10)
data['rmdecade'] = 10 * (data['YearRemodAdd'] // 10)
data.head()
```

```python
fig = plt.figure(figsize=(11,8))
property_price = data.pivot_table('rmdecade', index='SalePrice',
columns='YearBuilt')
property_price.plot(figsize=(11,8))
plt.gca().set_xticklabels(['G2LivArea', 'decade', 'FullBath', 'TotRmsAbvGrd', 'GarageArea', 'ExterQual_TA', 'Foundation_PConc'])
plt.ylabel('Property Price')
```

```python
fig = plt.figure(figsize=(2,10))
fig = sns.boxplot(data.rmdecade)
```
