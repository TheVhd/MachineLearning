```python
x = hbble['recession_velocity'].values.reshape(-1, 1)
y = hbble['distance'].values.reshape(-1, 1)
```

```python
print(hbble['recession_velocity'].values)
print(hbble['recession_velocity'].values.shape)
print(hbble['distance'].values)
```

```python
print(x.shape, y.shape)
print(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
SEED = 42
"""
Note: The seed can be any integer, and 
is used as the seed for the random sampler. 
The seed is usually random, netting different results. 
However, if you set it manually, the sampler will return the same results. 
It's convention to use 42 as the seed as a reference to the popular novel 
series "The Hitchhiker’s Guide to the Galaxy".
"""
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = SEED)
```

```python
print(x_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
def calc(slope, recession_velocity_y, distance_x):
    return slope*recession_velocity_y+distance_x

score = calc(regressor.coef_, regressor.intercept_, 9.5)
print(score) 
```

```python
score = regressor.predict([[9.5]])
print(score)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
df_preds = pd.DataFrame({'Actual': y_test.squeeze(), 'Predicted': y_pred.squeeze()})
print(df_preds)
```

```python
X = hbble.distance.values.reshape(-1,1)
Y = hbble.recession_velocity
lr = LinearRegression()
lr.fit(X,Y)
```

```python
BIGGER_SIZE = 10
plt.rc('font', size=BIGGER_SIZE) # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE) # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE) # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE) # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE) # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE) # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title
plt.figure()
plt.title("Data from Hubble")
plt.scatter(X, Y, color='red', label="data")
plt.plot(X, lr.predict(X), color='blue', linewidth=1, label="fitted line")
plt.xlabel("Distance")
plt.ylabel("Recession velocity")
plt.legend()
plt.show()
```
